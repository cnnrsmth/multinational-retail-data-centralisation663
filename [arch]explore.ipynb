```python
import re

# Assume data_copy2 is already defined
date_test = pd.DataFrame({
    'opening_date': [
        '1996-10-25', '2013-04-12', '2014-01-02', '2019-09-09', '1995-05-15',
        '2001-01-04', '2000-06-01', '2022-09-05', '2004-09-11', 'October 2012 08',
        'July 2015 14', 'ZCXWWKF45G', '2022/01/20', '2020 February 01', 'May 2003 27',
        '7AHXLXIUEF', '2016 November 25', '2008/12/07', 'NULL', '0OLAK2I6NS',
        'October 2006 04', '2001 May 04', 'A3PMVM800J', '1994 November 24', 'February 2009 28',
        'GMMB02LA9V', 'March 2015 02', 'NULL', '13PIY8GD1H', 'NULL', '36IIMAQD58'
    ]
})

# Custom function to parse non-standard dates
def parse_non_standard_dates(date_str):
    try:
        if re.match(r'\b\w+ \d{4} \d{2}\b', date_str):  # e.g., 'October 2012 08'
            return pd.to_datetime(date_str, format='%B %Y %d', errors='coerce')
        elif re.match(r'\b\d{4}/\d{2}/\d{2}\b', date_str):  # e.g., '2022/01/20'
            return pd.to_datetime(date_str, format='%Y/%m/%d', errors='coerce')
        elif re.match(r'\b\d{4} \w+ \d{2}\b', date_str):  # e.g., '2020 February 01'
            return pd.to_datetime(date_str, format='%Y %B %d', errors='coerce')
        elif re.match(r'\b\w+ \d{4} \d{2}\b', date_str):  # e.g., 'May 2003 27'
            return pd.to_datetime(date_str, format='%B %Y %d', errors='coerce')
        elif re.match(r'\b\w+ \d{2}\b', date_str):  # e.g., 'October 2006 04'
            return pd.to_datetime(date_str, format='%B %d', errors='coerce').replace(year=2020)  # Assuming current year
        elif re.match(r'\b\w+ \d{4} \d{2}\b', date_str):  # e.g., 'March 2015 02'
            return pd.to_datetime(date_str, format='%B %Y %d', errors='coerce')
        else:
            return pd.to_datetime(date_str, errors='coerce')  # Use default coercion for others
    except Exception:
        return np.nan

# Apply custom parsing to the non-standard dates
date_test['opening_date'] = date_test['opening_date'].apply(lambda x: parse_non_standard_dates(x) if pd.isna(pd.to_datetime(x, errors='coerce')) else pd.to_datetime(x, errors='coerce'))

# Verify the conversion again
non_standard_entries_after = date_test[date_test['opening_date'].isna()]['opening_date']
print("Entries converted to NaT (True errors):", non_standard_entries_after)

# Standardize NaT to NaN
date_test['opening_date'] = date_test['opening_date'].where(pd.notnull(date_test['opening_date']), None)

# Display the result
display(date_test)
```

```python
import re

# Function to clean phone numbers
def clean_phone_number(phone):
    if pd.isna(phone):
        return np.nan
    # Remove non-digit characters
    phone = re.sub(r'\D', '', phone)
    # Add a plus sign at the start if the number starts with the country code
    if phone.startswith('00'):
        phone = '+' + phone[2:]
    elif not phone.startswith('+'):
        phone = '+' + phone
    return phone

# Apply the cleaning function to the phone_number column
data_users_copy['phone_number'] = data_users_copy['phone_number'].apply(clean_phone_number)

# Display unique values after cleaning
phone_number_unique_cleaned = data_users_copy['phone_number'].unique()
print("Cleaned phone_number unique values:", phone_number_unique_cleaned[:10])
```
