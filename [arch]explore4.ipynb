```python
# Function to print unique values for each column in a dataframe
def print_unique_values(df):
    for column in df.columns:
        unique_values = df[column].unique()
        print(f"Unique values for column '{column}':")
        print(unique_values)
        print("\n")

# Apply the cleaning methods
cleaned_df = data_cleaning.clean_store_details(stores_df)

# Print unique values for each column in the cleaned dataframe
print("Unique values for each column after cleaning:")
print_unique_values(cleaned_df)
```
